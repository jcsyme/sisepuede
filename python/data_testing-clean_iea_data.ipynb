```python
all_countries = set(df_read["Country"])
yrs = [2017, 2018, 2019]
country = "Mexico"

# TEMPORARY


df_tmp = df_read[df_read["Country"] == country]
df_tmp = df_tmp[["Product", "Flow"] + yrs]
dict_repl = {"Nuclear": "Electricity", "Renewables and waste": "Electricity"}
df_tmp["Product"].replace(dict_repl, inplace = True)

dict_agg = {"Product": "first", "Flow": "first"}
dict_agg.update(dict(zip(yrs, ["sum" for x in yrs])))
df_tmp = df_tmp.groupby(["Product", "Flow"]).agg(dict_agg).reset_index(drop = True)

vec_pop = np.array(df_pop[df_pop["Country Name"] == country][[str(x) for x in yrs]])
vec_gdp = np.array(df_gdp[df_gdp["Country Name"] == country][[str(x) for x in yrs]])/(10**9)
# array
array_div = np.ones((len(keep_flows), len(yrs)))
array_div[0,:] = vec_pop
array_div[1,:] = vec_gdp
array_div[2,:] = vec_gdp
df_div = pd.DataFrame(array_div, columns = [str(x) + "_div" for x in yrs])
keep_flows = ["Residential (PJ)", "Commercial and public services (PJ)", "Other final consumption (PJ)"]
df_div["Flow"] = keep_flows
# rename for values
df_tmp = pd.merge(df_tmp, df_div, how = "left")
df_tmp = df_tmp.rename(columns = dict(zip(yrs, [str(x) + "_val" for x in yrs])))

array_vals_norm = np.array(df_tmp[[str(x) + "_val" for x in yrs]])/np.array(df_tmp[[str(x) + "_div" for x in yrs]])
df_vals_norm = pd.DataFrame(array_vals_norm, columns = [str(x) for x in yrs])
df_tmp = pd.concat([df_tmp, df_vals_norm], axis = 1)
df_tmp["mean"] = df_tmp[[str(x) for x in yrs]].apply(np.mean, axis = 1)
df_tmp = df_tmp[df_tmp["Flow"].isin(keep_flows)]


df_tmp_not_total = df_tmp[~df_tmp["Product"].isin(["Total"])].reset_index(drop = True).rename(columns = {"mean": "value"})
df_tmp_total = df_tmp[df_tmp["Product"].isin(["Total"])].reset_index(drop = True).rename(columns = {"mean": "total"})
df_tmp_compare = pd.merge(df_tmp_not_total[["Product", "Flow", "value"]], df_tmp_total[["Flow", "total"]], how = "left")
df_tmp_compare["proportion"] = np.array(df_tmp_compare["value"])/np.array(df_tmp_compare["total"])



pop = int(df_pop[df_pop["Country Name"] == "Costa Rica"]["2019"].iloc[0])#5046057
gdp = float(df_gdp[df_gdp["Country Name"] == "Costa Rica"]["2019"].iloc[0])/(10**9)#63.43013
vec_dat = [pop, gdp, gdp]
df_dat = pd.DataFrame({"Flow": keep_flows, "cr": vec_dat})

df_cr = pd.merge(df_tmp_compare, df_dat)
df_cr["est"] = np.array(df_cr["value"])#*np.array(df_cr["cr"])
df_cr["est_total"] = np.array(df_cr["total"])#*np.array(df_cr["cr"])

# assume some proprtion of electric demand that is used for heat energy (cooking, heating the home, etc.)
prop_elec_heat = 0.3

df_cr_new = []

fld_to_use = "est"
for flow in keep_flows:
    
    df_cr_tmp = df_cr[df_cr["Flow"] == flow]
    elec = float(df_cr_tmp[df_cr_tmp["Product"] == "Electricity"][fld_to_use].iloc[0])#*prop_elec_heat
    elec_heat = elec*prop_elec_heat
    elec_appl = elec - elec_heat
    
    vec_out_nm = ["Electricity - Appliance", "Electricity - Heat"]
    vec_out_dat1 = [elec_appl, elec_heat]
    
    df_filt = df_cr_tmp[df_cr_tmp["Product"] != "Electricity"]
    non_elec = np.array(df_filt[fld_to_use])
    non_elec_nms = df_filt["Product"]
    heat_ener_total = np.sum(non_elec) + elec_heat
    
    vec_out_nm += list(non_elec_nms)
    vec_out_dat1 += list(df_filt[fld_to_use])
    
    vec_out_dat2 = [elec_appl] + [heat_ener_total for x in range(len(df_filt) + 1)]
    
    df_cr_new_app = pd.DataFrame({"Flow": [flow for x in vec_out_dat2], "Product": vec_out_nm, "num": vec_out_dat1, "den": vec_out_dat2})
    df_cr_new.append(df_cr_new_app)
    

df_cr_new = pd.concat(df_cr_new, axis = 0).reset_index(drop = True)
df_cr_new["props"] = np.array(df_cr_new["num"])/np.array(df_cr_new["den"])


df_1 = pd.DataFrame({"Product": ["Oil products"], "pd": [2]})
df_cr_new = pd.merge(df_cr_new, df_1, how = "left")
df_cr_new["pd"].fillna(1, inplace = True)
df_cr_new["props_adj"] = np.round(np.array(df_cr_new["props"])/np.array(df_cr_new["pd"]), 3)
```
